```python
   classes=['paython','cloud','AI', 'java','c++']
    
class_1=[22,34,32,24,54]

class_2=[67,65,43,23,12]

class_3=[77,68,49,28,92]

plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.5

plt.bar(classes_index,class_1 ,width,color='y')

plt.bar(classes_index+width,class_2 ,width,color='r')

plt.xticks(classes_index+width,classes)

plt.bar(classes_index+width+width+width,class_3 ,width,color='b')

plt.title('Bar chart for students', fontsize=20)

plt.xlabel('Subjects',fontsize=15)

plt.ylabel('students',fontsize=15)

style.use('ggplot')
plt.savefig("D:\\Bar_chat.jpg")
plt.show()
```
