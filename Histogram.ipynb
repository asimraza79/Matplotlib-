```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
Cs_students_age=np.random.randint(15,40 ,(100))
Is_students_age=np.random.randint(20,40 ,(100))
```

```python
print(Cs_students_age)
print(Is_students_age)
```

```python
plt.hist(Cs_students_age)
plt.title("Cs Students")
plt.show()
```

```python
plt.hist(Cs_students_age,orientation='horizontal',rwidth=0.8,stacked=True)
plt.title("Cs Students")
plt.show()
```

```python
bins= [15,20,25,30,35,40,45]
plt.hist(Cs_students_age, bins ,rwidth=0.8 , color='r')
plt.title("Cs Students")
plt.show()
```

```python
bins= [15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([Cs_students_age, Is_students_age] ,bins ,rwidth=0.8 , color=['b', 'y'],histtype='bar' ,label=['cs student', 'Is students'])
#plt.hist(Is_students_age, bins ,rwidth=0.8 , color='y',histtype='bar' ,label='Is student')
plt.title("Cs&Is Students", fontsize=10)
plt.xlabel('studnets')
plt.ylabel('age')
plt.legend(loc=4)
plt.show()
```
