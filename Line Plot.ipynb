```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
temperature=[40,35.5,41,36.9,39.7,40.2,41]
plt.plot(day , temperature)
plt.title("Ryk Temperature")
plt.xlabel('days')
plt.ylabel('temperature')
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
temperature1=[40,35.5,41,36.9,39.7,40.2,41]
temperature2=[40,45,46,41,39,46.9,40.7]
plt.title('ryk & sdk temperature')
plt.xlabel('ryk')
plt.ylabel('sdk')
plt.plot(day, temperature1, 'yo--',linewidth=4, markersize=10, label= ('Ryk temp'))
plt.plot(day, temperature2, 'bo--',linewidth=4, markersize=10, label= ('sdk temp') )
plt.legend()
plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
temperature=[40,35.5,41,36.9,39.7,40.2,41]
plt.plot(day , temperature, 'go--', linewidth=3 ,markersize=10)
plt.title("Ryk Temperature")
plt.xlabel('days')
plt.ylabel('temperature')
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
temperature=[40,35.5,41,36.9,39.7,40.2,41]
plt.plot(day , temperature, 'go--', linewidth=3 ,markersize=10 , label='temp line')
plt.title("Ryk Temperature" ,fontsize=20)
plt.legend( loc=4 )
plt.xlabel('days' , fontsize=20) 
plt.ylabel('temperature', fontsize=20)
plt.grid()
plt.show()
```

# Add Two cities

```python
from matplotlib import style
day=[1,2,3,4,5,6,7]
Ryktemperature=[40,35.5,41,36.9,39.7,40.2,41]
sdktemperature=[35.9,36.5,41.7,45,42,43,41.5]
style.use('ggplot')
plt.plot(day , Ryktemperature, 'go--', linewidth=3 ,markersize=10, label='Ryktem')
plt.plot(day , sdktemperature, 'ro:', linewidth=3 ,markersize=10, label='sdktem')
plt.title("Ryk & sdk Temperature" ,fontsize=20)
plt.legend(loc=4 )
plt.xlabel('days' , fontsize=20) 
plt.ylabel('temperature', fontsize=20)
plt.grid(color='w', linestyle=':', linewidth=1)
plt.show()
```
