```python
import matplotlib.pyplot as plt
```

```python
classes=['java','c','paython','bio','ml']
class_1=[23,26,43,34,42]
plt.pie([1])
plt.show()
```

```python
classes=['java','c','paython','bio','Ml']
class_1=[23,26,43,34,42]
explode=[0,0.2,0,0.05,0]
textprops={'fontsize':15}
colors=['r','m','y','b','g']
plt.pie(class_1, labels =classes,explode=explode,colors=colors,textprops=textprops,shadow=True,radius=2,
        labeldistance=1.1,normalize=True,autopct="%0.2f%%")
#plt.legend()
plt.show()
```

```python
classes=['java','c','paython','bio','Ml']
class_1=[23,26,43,34,42]
explode=[0,0.2,0,0.05,0]
textprops={'fontsize':15}
colors=['r','m','y','b','g']
plt.figure(figsize=(10,5))
wedgeprops={'linewidth': 4,'edgecolor':'k','width':2}
plt.pie(class_1, labels =classes,explode=explode,colors=colors,textprops=textprops,shadow=True,radius=2,
        labeldistance=1.6,normalize=True,autopct="%0.1f%%",startangle=270,pctdistance=0.8, wedgeprops= wedgeprops,  
        frame=True,rotatelabels=True, center=(2, 3),counterclock=True,)
#plt.legend()
plt.show()
```
